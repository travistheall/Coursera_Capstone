```python
import pandas as pd
from pandas import json_normalize

import numpy as np

import requests

import random
from geopy.geocoders import Nominatim
from IPython.display import Image 
from IPython.core.display import HTML     
import folium

from sklearn.cluster import KMeans 
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
neighCenterData = pd.read_excel('NeighborhoodMaxMinCenter.xlsx')
```

```python
geojson = requests.get('https://opendata.arcgis.com/datasets/7c5e82ef83834de2ad2478efc86744ae_0.geojson').json()
df = json_normalize(geojson["features"])
coords = 'geometry.coordinates'
brgeodf = (df[coords].apply(lambda r: [(i[0],i[1]) for i in r[0]])
           .apply(pd.Series).stack()
           .reset_index(level=1).rename(columns={0:coords,"level_1":"point"})
           .join(df.drop(coords,1), how='left')).reset_index(level=0)
brgeodf[['Latitude', 'Longitude']] = brgeodf[coords].apply(pd.Series)
```

```python
brgeodf.rename(columns={'geometry.coordinates':'Coordinates',
                        'properties.ID':'ID',
                        'properties.NEIGHBORHOOD':'Neighborhood',
                        'properties.COMMUNITY':'Community',
                        'properties.POPULATION':'Population',
                        'properties.PLANNING_DISTRICT':'PanningDistrict',
                        'geometry.type':'Geometry'}, 
                 inplace=True)
```

```python
address = 'IBM, 100 North St, Baton Rouge, LA 70802'

geolocator = Nominatim(user_agent="IBM Explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
```

```python
map_br = folium.Map(location=[latitude, longitude], zoom_start=13)
map_br
```

```python
for lat, long, neighborhood, community in zip(brgeodf['Latitude'],
                                              brgeodf['Longitude'],
                                              brgeodf['Neighborhood'],
                                              brgeodf['Community']):
    label = '{}, {}'.format(neighborhood, community)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, long],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_br)
```

```python
averageCoord = {}

for x in range(len(df)):
    lat = []
    long = []
    coords = df.iloc[x][-1][0]
    for coord in coords:
        lat.append(coord[0])
        long.append(coord[1])
    
    latarray = np.array(lat)
    longarray = np.array(long)
    latavg = np.average(latarray)
    longavg = np.average(longarray)
    averageCoord[df.iloc[x][2]]=[latavg, longavg]
    
avgCoordDf = pd.DataFrame.from_dict(data=averageCoord)
```

```python
kMeansX = []

for x in brgeodf['Coordinates']:
    kMeansX.append(list(x))
    
kMeansX = np.array(kMeansX)
k_means = KMeans(init="k-means++", n_clusters=number_of_neighborhoods, n_init=12)
kMeansY = brgeodf['Neighborhood']
k_means.fit(kMeansX)
k_means_labels = k_means.labels_
k_means_cluster_centers = k_means.cluster_centers_
fig = plt.figure(figsize=(15, 10))
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means_labels))))
ax = fig.add_subplot(1, 1, 1)

for k, col in zip(range(len(k_means_cluster_centers)), colors):
    my_members = (k_means_labels == k)
    cluster_center = k_means_cluster_centers[k]
    ax.plot(kMeansX[my_members, 0], kMeansX[my_members, 1], 'w', markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=6)


ax.set_title('KMeans')
ax.set_xticks(())
ax.set_yticks(())
for x in avgCoordDf.columns:
    plt.plot(avgCoordDf[x][0], avgCoordDf[x][1], 'ro')
plt.show()
```
