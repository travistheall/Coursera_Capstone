```python
def getNearbyVenues(names, radius, latitudes, longitudes):
    
    venues_list=[]
    for name, radius, lat, lng in zip(names, radius, latitudes, longitudes):           
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# one hot encoding
br_onehot = pd.get_dummies(br_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
br_onehot['Neighborhood'] = br_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [br_onehot.columns[-1]] + list(br_onehot.columns[:-1])
br_onehot = br_onehot[fixed_columns]

print(br_onehot.shape)
```

```python
br_grouped = br_onehot.groupby('Neighborhood').mean().reset_index()
print(br_grouped.shape)
br_grouped.to_excel('br_grouped.xlsx')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = br_grouped['Neighborhood']

for ind in np.arange(br_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(br_grouped.iloc[ind, :], num_top_venues)
```

```python
neighborhoods_venues_sorted.to_excel('neighborhoods_venues_sorted.xlsx')
```

```python
# set number of clusters
kclusters = 3

br_grouped_clustering = br_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(br_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
#neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
neighborhoods_venues_sorted['Cluster Labels'] = kmeans.labels_

br_merged = neighCenterData

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
br_merged = br_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

br_merged.head() # check the last columns!
```

```python
# create map
map_clusters = folium.Map(location=[brCenterLat, brCenterLong], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
br_merged_map = br_merged.dropna()
# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(br_merged_map['Center Latitude'],
                                  br_merged_map['Center Longitude'],
                                  br_merged_map['Neighborhood'],
                                  br_merged_map['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster-1)],
        fill=True,
        fill_color=rainbow[int(cluster-1)],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

```python
distortions = [] 
inertias = [] 
mapping1 = {} 
mapping2 = {} 
K = range(1,10)
  
for k in K: 
    kmeanModel = KMeans(n_clusters=k).fit(br_grouped_clustering) 
    kmeanModel.fit(br_grouped_clustering)     
      
    distortions.append(sum(np.min(cdist(br_grouped_clustering, kmeanModel.cluster_centers_, 
                      'euclidean'),axis=1)) / br_grouped_clustering.shape[0]) 
    inertias.append(kmeanModel.inertia_) 
  
    mapping1[k] = sum(np.min(cdist(br_grouped_clustering, kmeanModel.cluster_centers_, 
                 'euclidean'),axis=1)) / br_grouped_clustering.shape[0] 
    mapping2[k] = kmeanModel.inertia_ 
```

```python
plt.plot(K, distortions, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Distortion') 
plt.title('The Elbow Method using Distortion') 
plt.show() 
```

```python
kmeans_per_k = [KMeans(n_clusters=k).fit(br_grouped_clustering) for k in range(1,10)]  
kmeans_per_k
silhouette_scores = [metrics.silhouette_score(br_grouped_clustering, model.labels_) for model in kmeans_per_k[1:]]
#silhouette_scores
#kmeans_per_k
```

```python
plt.plot(range(2,10), silhouette_scores, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Silhouette Scores') 
plt.title('Evaluation Using Silhouette Scores') 
plt.show() 
```
